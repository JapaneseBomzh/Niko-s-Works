```python
import numpy as np
import matplotlib.pyplot as plt
def f1(x,y):
    return x+np.sin(y/np.pi)
def f2(x,y):
    return x+np.cos(y/np.sqrt(2))
def f3(x,y):
    return np.sin(x)-np.cos(y)
x=[0,0.2,0.4,0.6,0.8,1]

def euler(f,x,y0,step):
    y=[];y.append(y0)
    for i in range (0,len(x)-1):
        y.append(y[i]+step*(f(x[i],y[i])))
        print(y)
    return y
y = euler(f3,x,1,0.2)


fig = plt.figure()
l1 = plt.plot(x, y)
plt.show()
```

```python
def euler_k(f,x,y0,step):
    y=[];y.append(y0)
    for i in range (0,len(x)-1):
        y.append(y[i]+step/2*(f(x[i],y[i])+f(x[i+1],(y[i]+step*(f(x[i],y[i]))))))
        print(y)
    return y
y = euler_k(f3,x,1,0.2)


fig = plt.figure()
l1 = plt.plot(x, y)
plt.show()
```
