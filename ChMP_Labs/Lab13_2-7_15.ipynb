```python
import numpy as np
import matplotlib.pyplot as plt

def f1(x,y):
    return x+np.sin(y/np.pi)
def f2(x,y):
    return x+np.cos(y/np.sqrt(2))
step=0.1

def x_create(x0,xn,step):
    x=[];x.append(x0);i=0
    xk=0
    while round(xk,2)<=xn:
        x.append(round((x[i]+step),2))
        i+=1;xk=round(x[i]+step,2)
    return x

x=x_create(1.7,2.7,step)
print(x)

def euler(f,x,y0,step):
    y=[];y.append(y0)
    for i in range (0,len(x)-1):
        y.append(round(y[i]+step*(f(x[i],y[i])),5))
    return y

y = euler(f1,x,5.3,step)
print(y)

plt.figure()
plt.plot(x, y)
plt.grid()
plt.show()
```

```python
x=x_create(0.8,1.8,step)
print(x)

def euler_k(f,x,y0,step):
    y=[];y.append(y0)
    for i in range (0,len(x)-1):
        y.append(round(y[i]+step/2*(f(x[i],y[i])+f(x[i+1],(y[i]+step*(f(x[i],y[i]))))),5))
    return y

y = euler_k(f2,x,1.4,step)
print(y)

plt.figure()
plt.plot(x, y)
plt.grid()
plt.show()
```
