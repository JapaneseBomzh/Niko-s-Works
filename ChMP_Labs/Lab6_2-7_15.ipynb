```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import lagrange


def Lagrange1(X,Y,m,n,x):
    S=0
    for i in range(0,n):
        d1=1
        d2=1
        for j in range(0,m):
            if i==j:
                d1*=(x-X[j])
                d2*=(X[i]-X[j])
        S+=(d1/d2)*Y[i]
    return S
x=np.array([-4.,-3.,-1,3,-3.5,-2,1.5,2], dtype=float)
y=np.array([-15.,5.,3.,-1.,-2.63,9,-16.38,-15], dtype=float)
xnew=np.linspace(np.min(x),np.max(x),100) 
ynew=[Lagrange1(x,y,len(x),len(y),i) for i in xnew]

plt.plot(xnew,ynew, x, y, 'bo')
plt.title('Lagrange Polynomial_1')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
f = lagrange(x, y)

plt.plot(xnew, f(xnew), 'g', x[range(0,4)], y[range(0,4)], 'ro',x[range(4,8)], y[range(4,8)],'k*')
plt.title('Lagrange Polynomial_2')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
