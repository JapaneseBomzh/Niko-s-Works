```python
import matplotlib.pyplot as plt
from scipy.interpolate import CubicSpline
import numpy as np
x= [0.9,1.2,1.6,2.1,2.8]
y= [3.47,4.53,2.86,1.64,2.25]

x1 = [0.9,1,1.2,1.4,1.6,1.8,2,2.1,2.2,2.4,2.6,2.8]
y1 = [3.47,3.985803442,4.53,3.937355637,2.86,2.094910576,1.716895984,1.64,1.621065441,1.721912945,1.953317267,2.25]

spl = CubicSpline(x1, y1)
xs = np.linspace(0.9, 2.8, 1000)

plt.plot(x,y,"bo",x1,y1,'r*',xs, spl(xs), 'g', lw=2.5)

plt.show()
```
