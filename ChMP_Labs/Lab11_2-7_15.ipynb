```python
import numpy as np 
import matplotlib.pyplot as plt 


def f(x): 
  return np.cos(2*x)+2*x

x = np.array ([i*0.1 for i in range(1, 11 )]) 
y = np.array([f(x)])
print ('x=',x)
print ('y=',y)

mean_x = np.mean(x) 
mean_y = np.mean(y) 

mean_x2 = np.mean(x**2)
mean_xy = np.mean (x*y)

print('mean_x=', mean_x, '\n', 'mean_y=', mean_y, '\n', 'mean_xy=', mean_xy, '\n', 'mean_x2=',mean_x2)

a1 = (mean_xy - mean_x*mean_y)/(mean_x2*(np.mean(x))**2)
a0 = mean_y - (a1* mean_x) 

  

print('Coefficients', 'a=', round(a0,2), 'b=', round(a1,2)) 



plt.figure(figsize=(8, 6))
plt.title("M N K")
plt.xlabel("x")
plt.ylabel("y")

plt.autoscale(tight=True)


plt.scatter(x, y)

fx = np.linspace(x[0], x[-1] , 1000)

plt.plot(fx, f(fx), linewidth=2)
plt.plot(x, a1*x + a0, 'r', label='Fitted line') 
plt.grid()
plt.savefig('data.png', dpi=50)
plt.show()
```
