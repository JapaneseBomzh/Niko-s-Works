```python
import numpy as np
import matplotlib.pyplot as plt
def f(x):
    return 3*pow(x,4)-10*pow(x,3)+pow(x,2)-5*x-3

a=-0.5
b=4
eps=0.0001
def method_1(a,b,eps):
    while abs(f(b)-f(a))>eps:
        c=(a+b)/2
        if f(a)*f(c)<0:
            b=c
        else:
            a=c
    x=c
    print(x)
method_1(a,b,eps) # щоб визначити перший корінь (від А до Б)
method_1(b,a,eps) # щоб визначити другий корінь (від Б до А)
x = np.arange(-2, 4, 0.01)

plt.plot(x, f(x))

plt.xlabel('x')

plt.ylabel('f(x)')

plt.title('The halving method')
plt.annotate('local min', xy=(5/2, f(5/2)), xytext=(3,-45 ),
             arrowprops=dict(facecolor='black', shrink=0.05),
             )
plt.grid()
plt.text(0,75,r'$f(x) = 3x^4-10x^3+x^2-5x-3$', fontsize=10, bbox={'facecolor':'yellow','alpha':0.2}) 

plt.show()
```
