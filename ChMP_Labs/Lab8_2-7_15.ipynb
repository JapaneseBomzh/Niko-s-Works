```python
def x():
    x=[3.50,3.55,3.60,3.65,3.70,3.75,3.80,3.85,3.90,3.95,4.00]
    return x
def y():
    y=[33.1154,34.8133,36.5982,38.4747,40.4473,42.5211,44.7012,46.9931,49.4024,51.9354,54.5982]
    return y
x1_1=3.522
x2_1=3.643
x3_1=3.675
n=11

def newton_1(n,X,Y,x):
    
    Xl=Y
    delta=[0 for i in range(n)]
    delta[0]=Xl[0]
    for i in range(n-1):
        for j in range(n-1):
            Xl[j]=Xl[j+1]-Xl[j]
        delta[i+1]=Xl[0]
    h=X[1]-X[0]
    q=(x-X[0])/h
    S=delta[0]
    fact=1
    d=1
    for i in range(1,n):
        fact=fact*i
        d=d*(q-i+1)
        S=S+(delta[i]*d)/fact
    return S

y1_1=newton_1(n,x(),y(),x1_1)
y2_1=newton_1(n,x(),y(),x2_1)
y3_1=newton_1(n,x(),y(),x3_1)
print(y1_1)
print(y2_1)
print(y3_1)
```

```python

x1_2=3.905
x2_2=4.005
x3_2=3.852
def newton_2(n,X,Y,x):
    delta=[0 for i in range(n)]
    Xl=Y 
    delta[0]=Xl[n-1]
    for i in range(n-1):
        for j in range(n-i-1):
            Xl[j]=Xl[j+1]-Xl[j]
        delta[i]=Xl[n-i-1]
    h=X[1]-X[0]
    q=(x-X[n-1])/h
    S=delta[0]
    fact=1
    d=1
    for i in range(1,n):
        fact*=i
        d*=(q+(i-1))
        S+=(delta[i]*d)/fact
    return S   

y1_2=newton_2(n,x(),y(),x1_2)
y2_2=newton_2(n,x(),y(),x2_2)
y3_2=newton_2(n,x(),y(),x3_2)
print(y1_2)
print(y2_2)
print(y3_2)
```
